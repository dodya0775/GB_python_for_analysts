```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep = ';')
df.head()
```

```python
print('\ntask_2')
print('2. На основе переменной quality с оценками вина создать новую переменную good: если quality > 5, то 1, иначе 0.')
print('-' * 77)
```

```python
df['good'] = df['quality'].apply(lambda x: 'yes' if x > 5 else 'no')
df.head()
```

```python
#Fivethirtyeight() или 538 стиль построения создает красивые графики с холодными цветами и толстыми весовыми линиями. 
plt.style.use('fivethirtyeight') 
```

```python
# Готовим данные для графика в виде сводной таблицы
data1 = df.groupby('quality')['fixed acidity'].agg(fixed_acidity = 'mean').reset_index().sort_values(by = 'quality', ascending = False)  
data1.head(7)
```

```python
# 1.Создаем линейный график

plt.figure(figsize = (7,5))

plt.plot(data1['fixed_acidity'], data1['quality'])

plt.title('Влияние фиксированной кислотности на вкус')
plt.xlabel('Фиксированная кислотность')
plt.ylabel('Оценка')

plt.show()
```

```python
print('Вывод: оценка потребителей становится выше при снижении кислотности')
```

```python
# Готовим данные для графика в виде сводной таблицы
data2 = df.groupby('quality')['residual sugar'].agg(residual_sugar = 'mean').reset_index().sort_values(by = 'quality', ascending = False)  
data2.head(10)
```

```python
# 2.Строим столбчатую диаграмму

plt.figure(figsize = (7,5))

plt.bar(data2['quality'],data2['residual_sugar'])

plt.title('Влияние остаточного сахара на вкус')
plt.xlabel('Оценка')
plt.ylabel('Уровень остаточного сахара')
plt.xticks(rotation = 30)

plt.show()
```

```python
print('Вывод: максимальная оценка потребителей достигается при уровне остаточного сахара 4')
```

```python
# Готовим данные для графика в виде сводной таблицы
data3 = df.groupby('quality')['alcohol'].agg(alcohol = 'mean').reset_index().sort_values(by = 'quality', ascending = False)  
data3.head()
```

```python
# 3. Строим диаграмму рассеяния

plt.figure(figsize = (7,5))

plt.scatter(data3['alcohol'], data3['quality'])

plt.title('Влияние алкоголя на вкус')
plt.xlabel('Градусы')
plt.ylabel('Оценка')

plt.show()
```

```python
print('Вывод: максимальная оценка потребителей достигается при увеличении содержания алкоголя :)')
```

```python
# Готовим данные для графика
data4 = df['good'].value_counts()
data4.head()
```

```python
# 4. Строим круговую диаграмму

plt.figure(figsize = (7,5))

colors = ['yellow','green']

plt.pie(data4, colors=colors, autopct='%1.1f%%')

plt.title('Соотношение хороших и плохих вин')
plt.legend(data4.index)
plt.show()
```

```python
print('Вывод: хороших вин больше в два раза')
```

```python
# Строим корреляциогнную матрицу
correlation = df.corr()
correlation
```

```python
# 5. Строим хитмэп по матрице корреляций
plt.figure(figsize = (15,7))
sns.heatmap(correlation, cmap = 'Blues', annot = True)
plt.title('Корреляционаая матрица')
plt.show()
```

```python
print('Вывод: максимальная корреляционная зависимость проявилась между остаточным сахаром и плотностью')
```
